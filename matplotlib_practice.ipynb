## Line Chart

```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# create the pot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# create cell with longet ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
plt.plot(x_axis, y_axis, marker = "*", color = "blue", linewidth = 2, label = 'Boston')
plt.xlabel("Date")
plt.ylabel("fare($)")
plt.ylim(0, 45)
plt.title("PyBer Fare by Month")
plt.grid()
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker = "D", color = "green", linewidth = 2, label = "Boston")
ax.legend()
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
ax.set_title("PyBer Fare by Month")
ax.set_ylim(0, 45)
ax.grid()
```

## Scatter Plots and Bubble Charts

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, c= 'red', label = 'Chicago')
plt.legend()
plt.xlabel("Date")
plt.ylabel("Fare ($)")
plt.title("PyBer Fare by Month")
plt.xlim(0, 50)
plt.gca().invert_yaxis()
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

### using object oriented method

```python
fix, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c = 'xkcd:sky blue',s=[i*5 for i in y_axis], alpha = .8, edgecolor = 'k', linewidth = 2, label = "Boston")
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlim(0, 50)
ax.invert_yaxis()
```

## Pie Charts

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode = explode_values, labels = x_axis, autopct = '%.1f%%')
plt.show()
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

Object Oriented approach

```python
explode_values = [0, 0, 0.4, 0, 0, 0, 0.4, 0, 0, 0, 0, 0]
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
fig, ax = plt.subplots()
ax.pie(y_axis, labels = x_axis, autopct = '%.1f%%', radius = 2.85, explode = explode_values, shadow = True, 
      startangle = 95, counterclock = False, colors = colors)
plt.show()
```
